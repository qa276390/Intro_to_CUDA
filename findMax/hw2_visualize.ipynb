```python
import pandas as pd
import matplotlib. pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Output.csv')
```

```python
df
```

```python
np.log2(df[df.blocksPerGrid==10].threadsPerBlock)
```

```python
plt.plot( np.log2(df[df.blocksPerGrid==10].threadsPerBlock),df[df.blocksPerGrid==10].gputime, 'bo', label='gputime(sec)')
plt.plot( np.log2(df[df.blocksPerGrid==10].threadsPerBlock),df[df.blocksPerGrid==10]['diff']*1e9, 'r--', label='diff(*1e-9)')
plt.legend()
plt.xlabel('block size(2^x)')
print('x: threads per block, y: gpu time')
```
